#### List community area names and numbers with per capita income less than 11000.

```python
%sql SELECT name, type, length(type) FROM PRAGMA_TABLE_INFO('Chicago_Census_Data')
```

```python
%sql SELECT COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER, PER_CAPITA_INCOME FROM Chicago_Census_Data \
                                                                          WHERE PER_CAPITA_INCOME < 11000
```

#### List all case numbers for crimes involving minors

```python
%sql SELECT CASE_NUMBER FROM Chicago_Crime_Data \
                        WHERE DESCRIPTION LIKE '%minor%'
```

#### List all kidnapping crimes involving a child

```python
%sql SELECT * FROM Chicago_Crime_Data \
              WHERE DESCRIPTION LIKE '%CHILD%' AND PRIMARY_TYPE = "KIDNAPPING"
```

#### List the kind of crimes that were recorded at schools.


```python
%sql SELECT COUNT(DISTINCT(PRIMARY_TYPE)) FROM Chicago_Crime_Data
```

#### List the type of schools along with average safety score for each type

```python
%sql SELECT * FROM Chicago_Public_Schools LIMIT 5
```

```python
%sql SELECT name, type, length(type) FROM PRAGMA_TABLE_INFO('Chicago_Public_Schools')
```

```python
%sql SELECT "Elementary, Middle, or High School", AVG(SAFETY_SCORE) AS AVERAGE_SAFETY_SCORE FROM Chicago_Public_Schools \
                                                 GROUP BY "Elementary, Middle, or High School"
```

#### List 5 community areas with highest % of households below poverty line

```python
%sql SELECT COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER, PERCENT_HOUSEHOLDS_BELOW_POVERTY FROM Chicago_Census_Data \
                                                       ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 5
```

#### Which community area is most crime prone

```python
%sql SELECT COMMUNITY_AREA_NAME, CCD.COMMUNITY_AREA_NUMBER, COUNT(CASE_NUMBER) AS TOTAL_CRIME FROM Chicago_Census_Data CCD, Chicago_Crime_Data CCrD \
                                                       WHERE CCD.COMMUNITY_AREA_NUMBER = CCrD.COMMUNITY_AREA_NUMBER\
                                                       GROUP BY CCD.COMMUNITY_AREA_NUMBER \
                                                       ORDER BY TOTAL_CRIME DESC LIMIT 1
                                                       
```

#### Find the name of the community area with highest hardship index


```python
%sql SELECT COMMUNITY_AREA_NAME, HARDSHIP_INDEX FROM Chicago_Census_Data \
                                                WHERE HARDSHIP_INDEX = ( \
                                                SELECT MAX(HARDSHIP_INDEX) FROM Chicago_Census_Data )
```
